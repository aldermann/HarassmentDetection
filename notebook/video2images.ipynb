```
dataset = 'gdrive/My Drive/DA-20182/forest_dataset'
video_path = os.path.join(dataset, 'video', 'train')
image_path = os.path.join(dataset, 'image', 'train')
if not os.path.isdir(image_path):
  os.mkdir(image_path)
print(os.listdir(video_path))
for c in ['abnormal']:
  count = 0
  if not os.path.isdir(os.path.join(image_path, c)):
    os.mkdir(os.path.join(image_path, c))
  for v in os.listdir(os.path.join(video_path, c)):
    single_class_path = os.path.join(os.path.join(video_path, c, v))
    print(single_class_path)
    cap = cv2.VideoCapture(single_class_path)
    ret = True
    while ret:
        ret, frame = cap.read()
        if not ret:
          continue
        count += 1
        if count %25==0:
          try:
              cv2.imwrite(os.path.join(image_path, c, '%d.png'%count), frame)
          except Exception as e:
              print(str(e))
```
